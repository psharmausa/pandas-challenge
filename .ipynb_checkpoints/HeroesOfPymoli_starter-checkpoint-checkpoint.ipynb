```python
import pandas as pd

# File to Load (Remember to Change These)
data_file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(data_file)
df.head(100)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#get unique value of the Item name
UItem=df['Item Name'].nunique()

#average of price
average = df["Price"].mean()
average_2=str(round(average,2))

#total no. of purchase
TotalNumPurchase=df.count()[0]

#Total Revenue
total = df["Price"].sum()
total_2='$'+str(round((total),2))

# Create a DataFrame of frames using a dictionary of lists
summary_df = pd.DataFrame({ "Unique Item Name": [UItem],
    "Average of Price": [average_2],
    "Total No. of Purchase": [TotalNumPurchase],
    "Total Revenue": [total_2]})

#to hide index column
summary_df=summary_df.style.hide_index()
summary_df
```


## Purchasing Analysis (Gender)

```python
* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame
```

```python
# group by gender
ggrp = df.groupby(["Gender"])

#count of Purchase
cntp=ggrp["Item Name"].count()

# #average purchase
ap=round(ggrp["Price"].mean(),2)

# #total purchase
tpv=round(ggrp["Price"].sum(),2)


# # avg purchase per person
app=tpv/cntp



purchase_analysis_df  = pd.DataFrame()
purchase_analysis_df ["Item Name"]=cntp
purchase_analysis_df["Average Purchase Price"]= app                                   
purchase_analysis_df[ "Total Purchase Value"]= tpv                                     
purchase_analysis_df["Avg Total Purchase per Person"] = app
purchase_analysis_df["Total Purchase Value"] = purchase_analysis_df["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis_df["Average Purchase Price"]= purchase_analysis_df["Average Purchase Price"].map("${:,.2f}".format)
purchase_analysis_df["Avg Total Purchase per Person"] = purchase_analysis_df["Avg Total Purchase per Person"].map("${:,.2f}".format)
purchase_analysis_df


# first=df.groupby('Gender').agg({'Item Name':'count', 'Price': ['sum','mean']})
# first
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# print(df["Age"].max())
# print(df["Age"].min())
bins = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

df["Age Group"] = pd.cut(df["Age"],bins, labels=group_names)
df

#group
agrp = df.groupby("Age Group")

#total age count
ta = agrp["SN"].nunique()

#% by age
pba = (ta/ len(df["SN"].value_counts())) * 100

age_demog = pd.DataFrame({"Total Count": ta, "Percentage of Players": round(pba,2)})
age_demog.style.format({"Percentage of Players":"{:,.2f}%"})
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#purchase count
pcnt = agrp["Purchase ID"].count()

#average purchase price age
appa = agrp["Price"].mean()

#total purchase
tp = agrp["Price"].sum()

#averae purchase per person age
apppa=tp/ta

# data frame
age_demog = pd.DataFrame({"Purchase Count": pcnt,
                                 "Average Purchase Price": appa,
                                 "Total Purchase Value":tp,
                                 "Average Purchase Total per Person": apppa})

age_demog.style.format({"Average Purchase Price":"${:,.2f}",
                               "Total Purchase Value":"${:,.2f}",
                               "Average Purchase Total per Person":"${:,.2f}"})
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group by spender
spender = df.groupby("SN")

#purchase count spender
pcs= spender["Purchase ID"].count()

# average purchase price per spender
apps  = round( spender["Price"].mean(),2)

# purchase total spender
pts = spender["Price"].sum()

# dataframe
ts  = pd.DataFrame({"Purchase Count":  pcs,
                             "Average Purchase Price":  apps,
                             "Total Purchase Value":  pts})

ts.style.format({"Average Purchase Total":"${:,.2f}",
                                 "Average Purchase Price":"${:,.2f}", 
                                 "Total Purchase Value":"${:,.2f}"})

ts
#new sort by total purchase value
ns = ts.sort_values(["Total Purchase Value"], ascending=False).head(5)

ns.style.format({"Average Purchase Total":"${:,.2f}",
                                 "Average Purchase Price":"${:,.2f}", 
                                 "Total Purchase Value":"${:,.2f}"})

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#retrieve the item id, name and price
igrp = df[["Item ID", "Item Name", "Price"]]

#item group
item = igrp.groupby(["Item ID","Item Name"])

#purchase count item
pci = item["Price"].count()

#total purchase value
totpv  = (item["Price"].sum()) 

#item price
ip = totpv/pci

#most popular
mp  = pd.DataFrame({"Purchase Count":  pci, 
                                   "Item Price":  ip,
                                   "Total Purchase Value": totpv})
#most popular
mp.style.format({"Item Price":"${:,.2f}",
                                "Total Purchase Value":"${:,.2f}"})
#new popular
np = mp.sort_values(["Purchase Count"], ascending=False).head()
np
```
